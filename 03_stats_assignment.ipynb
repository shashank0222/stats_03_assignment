Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pylab
from scipy import stats
```

```python
arr = np.random.normal(loc = 4 , scale = 2 , size = 500)
```

```python
arr
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.histplot(arr , kde = True)

plt.subplot(1,2,2)
stats.probplot(arr,dist = "norm" , plot = pylab)

plt.show()
```

  formula to calculate pdf of a normal distribution =
  
  
    
    

```python
import math

def pdf(x,mean,std):
    var = float(std)**2
    denom = (2*math.pi*var)**0.5
    num = math.exp(-(float(x) - float(mean))**2/(2*var))
    return num/denom
```

```python
res = pdf(50,mean = np.mean(arr), std = np.std(arr))
```

```python
res
```

## Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Solution:

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, either Success or Failure. For example, if we toss a coin, there could be only two possible outcomes: heads or tails.

In binomial probability distribution, the number of â€˜Successâ€™ in a sequence of n experiments, where each time a question is asked for yes-no, then the boolean-valued outcome is represented either with success/yes/true/one (probability p) or failure/no/false/zero (probability q = 1 âˆ’ p). A single success/failure test is also called a Bernoulli trial or Bernoulli experiment, and a series of outcomes is called a Bernoulli process. For n = 1, i.e. a single experiment, the binomial distribution is a Bernoulli distribution. 


Properties : 

1. There are two possible outcomes: true or false, success or failure, yes or no.

2. There is â€˜nâ€™ number of independent trials or a fixed number of n times repeated trials.

3. The probability of success or failure remains the same for each trial.

4. Only the number of success is calculated out of n independent trials.

5. Every trial is an independent trial, which means the outcome of one trial does not affect the outcome of another trial.


Examples where binomial distribution can be applied : 

1. Taking a survey of positive and negative reviews from the public for any specific product or place.

2. By using the YES/ NO survey, we can check whether the number of persons views the particular channel.

3. To find the number of male and female employees in an organisation.

4. The number of votes collected by a candidate in an election is counted based on 0 or 1 probability.

 # Q7. How Binomial distribution different from Poisson distribution?

# Solution :

Binomial distribution only has two possible outcomes, 
whereas poisson distribution can have unlimited possible outcomes.

But for very large n and near-zero p binomial distribution is near identical to 
poisson distribution such that n * p is nearly equal to lam.

```python
arr = np.random.binomial(n = 1000 , p = 0.01  , size = 1000)
```

```python
arr1 = np.random.poisson(lam=10,size = 1000)
```
